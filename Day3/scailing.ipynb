```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(143)
```

```python
total_no=1000
class_0_ratio=0.9
n_class_0=int(total_no*class_0_ratio)
n_class_1=total_no - n_class_0
```

```python
class_0=pd.DataFrame({'feature1': np.random.normal(loc=0,scale=1,size=n_class_0),
                      'feature2': np.random.normal(loc=0,scale=1,size=n_class_0),
                      'target': [0]*n_class_0})
```

```python
class_0
```

```python
class_1=pd.DataFrame({'feature1': np.random.normal(loc=0,scale=1,size=n_class_1),
                      'feature2': np.random.normal(loc=0,scale=1,size=n_class_1),
                      'target': [1]* n_class_1})
```

```python
class_1
```

```python
df=pd.concat([class_1,class_0])
```

```python
df.reset_index(drop=True)
```

```python
from sklearn.utils import resample
```

```python
df_majority=df[df['target']==0]
df_minority=df[df['target']==1]
```

```python
df_minority_updated=resample(df_minority,n_samples=len(df_majority), replace=True,random_state=43)
```

```python
df_minority_updated.shape
```

```python
df=pd.concat([df_majority,df_minority_updated]).reset_index(drop=True)
```
