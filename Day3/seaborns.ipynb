```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x=iris.petal_length,y=iris.petal_width)
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
sns.displot(iris['petal_length'])
```

```python
sns.displot(iris['sepal_length'])
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x=tips.smoker,y=tips.tip)
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,hue='smoker',data=tips)
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,hue='time',data=tips,style='sex')
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='time',style='day')
```

```python
sns.catplot(x=tips.tip,y=tips.time,data=tips)
```

```python
sns.jointplot(x=tips.time,y=tips.tip)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(tips)
```
