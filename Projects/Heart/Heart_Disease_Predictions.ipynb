## IV. Train Test split

```python
from sklearn.model_selection import train_test_split

predictors = dataset.drop("target",axis=1)
target = dataset["target"]

X_train,X_test,Y_train,Y_test = train_test_split(predictors,target,test_size=0.20,random_state=0)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train,Y_train)

Y_pred_lr = lr.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Y_pred_lr))

from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred_lr))

from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,Y_pred_lr))
sns.heatmap(confusion_matrix(Y_test,Y_pred_lr),annot=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
import numpy as np

hypertune_logistic = LogisticRegression(max_iter=1000)

param_grid_logistic = {
    'C': np.logspace(-4, 4, 20),
    'solver': ['liblinear', 'lbfgs', 'saga', 'newton-cg']
}

random_search_logistic = RandomizedSearchCV(
    estimator=hypertune_logistic,
    param_distributions=param_grid_logistic,
    n_iter=20,
    scoring='accuracy',
    n_jobs=-1,
    cv=5,
    verbose=0,
    random_state=42
)

random_search_logistic.fit(X_train, Y_train)
best_hypertune_logistic = random_search_logistic.best_estimator_
Y_pred_hypertune_logistic = best_hypertune_logistic.predict(X_test)

accuracy_hypertune_logistic = accuracy_score(Y_test, Y_pred_hypertune_logistic)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# Scaling the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize Naive Bayes model
hypertune_naive_bayes = GaussianNB()

# Adjusting the range for var_smoothing to be more conservative
param_grid_naive_bayes = {
    'var_smoothing': np.logspace(-8, -4, 5)  # Narrowing the range
}

# Perform Randomized Search for hyperparameter tuning
random_search_naive_bayes = RandomizedSearchCV(
    estimator=hypertune_naive_bayes,
    param_distributions=param_grid_naive_bayes,
    n_iter=10,  # Number of iterations for random search
    scoring='accuracy',
    n_jobs=-1,
    cv=5,  # Increased cross-validation folds for more reliable results
    verbose=0,
    random_state=42
)

# Fit the model with the scaled data
random_search_naive_bayes.fit(X_train_scaled, Y_train)

# Get the best model after tuning
best_hypertune_naive_bayes = random_search_naive_bayes.best_estimator_

# Predict with the tuned model
Y_pred_hypertune_naive_bayes = best_hypertune_naive_bayes.predict(X_test_scaled)

# Calculate accuracy of the tuned model
accuracy_hypertune_naive_bayes = accuracy_score(Y_test, Y_pred_hypertune_naive_bayes)

# Compare the accuracy before and after tuning
baseline_accuracy = accuracy_score(Y_test, Y_pred_nb)
print(f"Baseline accuracy (Naive Bayes without tuning): {baseline_accuracy*100:.2f}%")
print(f"Tuned accuracy (Naive Bayes with hyperparameter tuning): {accuracy_hypertune_naive_bayes*100:.2f}%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier()

param_grid_knn = {
    'n_neighbors': np.arange(5, 31, 2),  # Range of neighbors, try odd numbers for better results
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2],
}

random_search_knn = RandomizedSearchCV(
    estimator=knn,
    param_distributions=param_grid_knn,
    n_iter=15,  # Increased iterations to explore more possibilities
    scoring='accuracy',
    n_jobs=-1,
    cv=5,  # Cross-validation
    verbose=0,
    random_state=42
)

random_search_knn.fit(X_train_scaled, Y_train)
best_knn = random_search_knn.best_estimator_
Y_pred_knn_tuned = best_knn.predict(X_test_scaled)

accuracy_knn_tuned = accuracy_score(Y_test, Y_pred_knn_tuned)
accuracy_knn = accuracy_score(Y_test, Y_pred_knn)

print(f"Baseline accuracy (KNN without tuning): {accuracy_knn*100:.2f}%")
print(f"Tuned accuracy (KNN with hyperparameter tuning): {accuracy_knn_tuned*100:.2f}%")

print("\nClassification Report (Tuned KNN):")
print(classification_report(Y_test, Y_pred_knn_tuned))

print("\nConfusion Matrix (Tuned KNN):")
cm_knn_tuned = confusion_matrix(Y_test, Y_pred_knn_tuned)
print(cm_knn_tuned)

sns.set(rc={'figure.figsize':(8, 6)})
sns.heatmap(cm_knn_tuned, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Confusion Matrix for Tuned KNN')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

scores = [accuracy_knn, accuracy_knn_tuned]
algorithms = ["KNN (Baseline)", "KNN (Tuned)"]

sns.set(rc={'figure.figsize':(8, 6)})
sns.barplot(x=algorithms, y=scores, palette="Blues_d")
plt.title('Accuracy Comparison: KNN (Baseline) vs. KNN (Tuned)')
plt.xlabel('Algorithms')
plt.ylabel('Accuracy Score')
plt.show()
```
