### EDA of algeiran Fire dataset 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Algerian_fire.csv',header=1)
```

```python
dataset.info()
```

```python
dataset.head()
```

### Cleaning the dataset

```python
dataset.isnull()
```

```python
dataset.isna().sum()
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
```

```python
dataset
```

```python
dataset=dataset.dropna().reset_index(drop=True)
```

```python
dataset.loc[[122]]
```

```python
dataset=dataset.drop(122).reset_index(drop=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.columns
```

```python
dataset.columns=dataset.columns.str.strip()
```

```python
dataset[["day","month","year","Temperature","RH","Ws","Region"]]=dataset[["day","month","year","Temperature","RH","Ws","Region"]].astype(int)
```

```python
Objects=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
```

```python
for i in Objects:
    if i !='Classes':
        dataset[i]=dataset[i].astype(float)
       
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.to_csv('cleaned_dataset.csv',index=False)
```

```python
df=dataset.drop(['day','month','year'],axis=1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df
```

```python
plt.style.use('seaborn')
df.hist(bins=50,figsize=(15,10))
plt.show()
```

```python
percentage=df['Classes'].value_counts(normalize=True)*100
```

```python
classeslabel=["Fire","Not Fire"]
plt.figure(figsize=(50,7))
plt.pie(percentage,labels=classeslabel,autopct="%1.1f%%")
plt.title("Percentage Chart")
plt.show()
```
