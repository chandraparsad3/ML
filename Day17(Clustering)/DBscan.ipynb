```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=900,noise=0.05)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.cluster import DBSCAN
```

```python
scan=DBSCAN(eps=0.5)
```

```python
scan.fit(X_scaled)
```

```python
scan.labels_
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=scan.labels_)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=y)
```
