```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM,Bidirectional
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow as tf
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
ps=PorterStemmer()
```

```python
voc_size=5000
sent_length=20
```

```python
nltk.download('stopwords')
```

```python
corpus = []
for i in range(0, len(message)):
    title = str(message['title'][i])  # Convert to string
    review = re.sub('[^a-zA-Z]', ' ', title)  # Use the 'title' variable
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repre=[one_hot(word,voc_size) for word in corpus]
```

```python
onehot_repre
```

```python
###Word Embedding
```

```python
embedding_docs=pad_sequences(onehot_repre,padding='pre',maxlen=sent_length)
```

```python
embedding_docs
```

```python
model.add(tf.keras.layers.Dense(128, activation='relu', trainable=True))
```

```python
###Model Trainging
V_features=40
model=Sequential()
model.add(Embedding(voc_size,V_features,input_length=sent_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(10))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

```

```python
X_final=np.array(embedding_docs)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_final,y_final,test_size=0.25,random_state=42)
```

```python
model.fit(X_train,y_train,validation_data=(X_final,y_final),epochs=10,batch_size=64)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
y_pred=model.predict(X_test)
```

```python
len(y_pred)
```

```python
y_pred=np.where(y_pred>0.6,1,0)
```

```python
len(y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
