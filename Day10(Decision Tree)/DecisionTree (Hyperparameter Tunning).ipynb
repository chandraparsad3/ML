### Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset=load_iris()
```

```python
dataset.DESCR
```

### Dependent and Independent Feature

```python
X=df.iloc[:,:-1]
y=dataset.target
```

```python
X,y
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Dtree=DecisionTreeClassifier()
```

```python
Dtree.fit(X_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(10,15))
tree.plot_tree(Dtree,filled=True)
```

### Post Prunning

```python
Dstree=DecisionTreeClassifier(max_depth=2)
Dstree.fit(X_train,y_train)
```

```python
tree.plot_tree(Dstree,filled=True)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
y_predict=Dstree.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predict))
print(classification_report(y_test,y_predict))
```

```python
Dttree=DecisionTreeClassifier(criterion='entropy',max_depth=2)
Dttree.fit(X_train,y_train)
```

```python
y_predict=Dttree.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predict))
print(classification_report(y_test,y_predict))
```

### Decision Tree with hyperparameter Tunning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={'criterion':['gini', 'entropy', 'log_loss'],
            'splitter': ['best','random'],
           'max_depth': [1,2,3,4,5],
           'max_features': ['auto','sqrt', 'log2']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
trees=DecisionTreeClassifier()
```

```python
DecisionTree_hyper=GridSearchCV(trees,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
DecisionTree_hyper.fit(X_train,y_train)
```

```python
y_predicts=DecisionTree_hyper.predict(X_test)
```

```python
score=accuracy_score(y_test,y_predicts)
```

```python
score
```

```python
cl_report=classification_report(y_predicts,y_test)
```
