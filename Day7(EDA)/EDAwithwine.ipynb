### EDA of red wine

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import csv

# Input and output file paths
input_csv = 'winequality-red.csv'
output_csv = 'redwine.csv'

# Read CSV file with semicolon delimiter and write to new CSV with comma delimiter
with open(input_csv, newline='', encoding='utf-8') as infile, \
     open(output_csv, 'w', newline='', encoding='utf-8') as outfile:
    
    # Setup CSV reader and writer
    csv_reader = csv.reader(infile, delimiter=';')
    csv_writer = csv.writer(outfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    # Read the header from input CSV
    header = next(csv_reader)
    # Write the header to output CSV
    csv_writer.writerow(header)
    
    # Iterate through each row in input CSV
    for row in csv_reader:
        # Process each row if needed
        # For example, you can modify the row or just write it directly to output CSV
        csv_writer.writerow(row)

print(f"Data has been successfully written to {output_csv}")
```
