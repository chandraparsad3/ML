```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=["A glass of milk",
      "A glass of water",
      "I am a good developer",
      "I am a good football player",
      "I am interested in ML and AI",
      "Shyam is good at coding mainly JAVA",
      ]
```

```python
sent
```

```python
### Defining Vocabulary Size
voc_size=10000
```

```python
onehot_repre=[one_hot(words,voc_size) for words in sent]
print(onehot_repre)
```

```python
### Word Embedding Techniques
import numpy as np
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
model=Sequential()
```

```python
embedding_docs=pad_sequences(onehot_repre,padding='pre',maxlen=8)
print(embedding_docs)
```

```python
model.add(Embedding(voc_size,10,input_length=8))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedding_docs))
```
